# Fig & Ax Properties

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot([1,2,3],[1,4,9])
ax2.plot([1,2,3],[1,2,3])
plt.show()
```

```python
fig = plt.figure(dpi=150)
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot([1,2,3],[1,4,9])
ax2.plot([1,2,3],[1,2,3])
plt.show()
```

## sharex, sharey

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot([1,2,3],[1,4,9])
ax2 = fig.add_subplot(122, sharey=ax1)
ax2.plot([1,2,3],[1,2,3])
plt.show()
```

```python
fig, axes = plt.subplots(1,2,sharey=True)
axes[0].plot([1,2,3], [1,4,9])
axes[1].plot([1,2,3], [1,2,3])
plt.show()
```

## squeeze flatten

```python
n,m = 2,3
fig,axes = plt.subplots(n,m,squeeze=False,figsize=(m*2,n*2))
idx = 0
for i in range(n):
    for j in range(m):
        axes[i][j].set_title(idx)
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])   
        idx+=1
        
plt.show()
```

```python
n,m = 2,3
fig,axes = plt.subplots(n,m,squeeze=False,figsize=(m*2,n*2))

for i in range(n):
    for j in range(m):
        axes[i][j].set_title(idx)
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])   
        idx+=1
        
plt.show()
```

## aspect

```python
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(121, aspect=1)
ax2 = fig.add_subplot(122, aspect=0.5)

ax1.set_ylim(0,1)
ax2.set_ylim(0,2)
plt.show()
```

# Gridspec

```python
fig = plt.figure(figsize=(8,5))
gs = fig.add_gridspec(3,3)

ax = [None for _ in range(5)]

ax[0] = fig.add_subplot(gs[0,:])
ax[0].set_title('gs[0, :]')

ax[1] = fig.add_subplot(gs[1,:-1])
ax[1].set_title('gs[1, :-1]')

ax[2] = fig.add_subplot(gs[1:, -1])
ax[2].set_title('gs[1:, -1]')

ax[3] = fig.add_subplot(gs[-1, 0])
ax[3].set_title('gs[-1, 0]')

ax[4] = fig.add_subplot(gs[-1, -2])
ax[4].set_title('gs[-1, -2]')

for ix in range(5):
    ax[ix].set_xticks([])
    ax[ix].set_yticks([])
    
plt.tight_layout()
plt.show()
```

## add_axes 

```python
fig = plt.figure(figsize=(8,5))
ax = [None for _ in range(3)]

ax[0] = fig.add_axes([0.1,0.1,0.8,0.4]) # x,y,dx,dy
ax[1] = fig.add_axes([0.15,0.6,0.25,0.6])
ax[2] = fig.add_axes([0.5,0.6,0.4,0.3])

for ix in range(3): 
    ax[ix].set_title(f'ax[{ix}]')
    ax[ix].set_xticks([])
    ax[ix].set_yticks([])
    
plt.show()
```

## insert_axes

```python
fig, ax = plt.subplots()
axin = ax.inset_axes([0.8,0.8,0.2,0.2])
plt.show()
```

```python
fig, ax = plt.subplots()

color = ['royalblue', 'tomato']
ax.bar(['A','B'], [1,2], color=color)

ax.margins(0.2)
axin = ax.inset_axes([0.8,0.8,0.2,0.2])
axin.pie([1,2], colors=color, autopct='%1.0f%%')
plt.show()
```

## make_axes_locatable

```python
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable

fig, ax = plt.subplots(1,1)
ax_divider = make_axes_locatable(ax)

ax = ax_divider.append_axes("right", size="7%", pad="2%")
plt.show()
```

```python
fig, ax = plt.subplots(1,1)

im = ax.imshow(np.arange(100).reshape((10,10)))

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)

fig.colorbar(im, cax=cax)
plt.show()
```
