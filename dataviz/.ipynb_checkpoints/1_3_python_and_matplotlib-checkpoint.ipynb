# 1. 왜 Matplotlib일까?

- numpy와 scipy를 베이스로 함(sklearn, pytorch, tensorflow, pandas ...)
- 다양한 시각화 방법론

    - 산점도

- Seaborn, Plotly, Bokeh, Altair 등 시각화 라이브러리 존재

# 2. 기본 Plot

## 2-1. Figure와 Axes

Figure = 틀   
Ax = 서브 플롯

```python
fig = plt.figure()
plt.show()
```

Figure은 틀이기 때문에 서브플롯을 최소 1개 이상 추가해야한다.

```python
fig = plt.figure()
ax1 = fig.add_subplot()
plt.show()
```

그래프의 사이즈는 figure의 사이즈로 서브플롯 ax의 사이즈를 조정한다 (inch 단위)   
notebook 환경에서는 크기보다는 비율로 생각하면 편함

```python
fig = plt.figure(figsize=(12,7))
#fig.set_facecolor('black')
ax = fig.add_subplot()
plt.show()
```

2개 이상 subplot을 그릴 때는 위치를 지정해주어야 한다.

```python
fig = plt.figure()
ax = fig.add_subplot(121) # y축 1칸, x축 2칸 일때 첫번째칸에 추가
ax = fig.add_subplot(122)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
plt.show()
```

```python
fig = plt.figure()
axes = []
for i in range(1,10):
    axes.append(fig.add_subplot(3,3,i))
plt.show()
print(axes)
```

## 2-3. 서브플롯 객체 ax에 그리기

### 객체지향 API

```python
fig = plt.figure()

x1 = [1,2,3]
x2 = [3,2,1]

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x1)
ax2.plot(x2)

plt.show()
```

# 3. Plot의 요소들 알아보기

## 3-1. 한 서브플롯에서 여러 개 그리기

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1]) # 파랑
ax.plot([1,2,3]) # 주황
ax.plot([3,3,3]) # 초록

plt.show
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,2,3], [1,2,3])
ax.bar([1,2,3], [1,2,3])

plt.show()
```

색을 명시해주는 것이 좋다

## 3-2. 색상 지정하기

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot([1,1,1], color='r')
ax.plot([2,2,2], color='forestgreen')
ax.plot([3,3,3], color='#000000')
plt.show()
```

## 3-3. 텍스트 사용하기

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')

plt.show()
```

### legend
label을 드러나게 하기 위해서는 legend로 추가해야한다.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()

plt.show()
```

### set_title
set_title로 제목을 추가할 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()
ax.set_title('visualization test')

plt.show()
```

## suptitle
suptitle로 전체 figure에 대한 제목을 지정할 수 있다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.set_title('ax1')
ax2.set_title('ax2')

fig.suptitle('figure') #super

plt.show()
```

### get_title
set_??으로 세팅한 정보는 get_??로 가져올 수있다.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()
ax.set_title('visualization test')

plt.show()
print('title of ax : ', ax.get_title())
```

### ticks
축에 적히는 수 위치를 지정

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()

ax.set_title('visualization test')
ax.set_xticks([0,1,2])

plt.show()
```

### ticklabels
축에 적히는 텍스트 수정

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()

ax.set_title('visualization test')
ax.set_xticks([0,1,2])
ax.set_xticklabels(['a','b','c'])
plt.show()
```

### text
일반적인 텍스트 추가 (위치에 따라)

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()

ax.set_title('visualization test')
ax.set_xticks([0,1,2])
ax.set_xticklabels(['a','b','c'])
ax.text(x=1, y=2, s='This is Text')

plt.show()
```

### annotate
텍스트 + 화살표 등 표시 방법 세팅 가능

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')
ax.legend()

ax.set_title('visualization test')
ax.set_xticks([0,1,2])
ax.set_xticklabels(['a','b','c'])
ax.annotate(text='This is Text', xy=(1,2),
            xytext = (1.2, 2.2),
            arrowprops = dict(facecolor='black'))

plt.show()
```
